### A common pitfall


````python
# Let’s initialize a row
row = [""] * 3
# Let’s make a board
board = [row] * 3
````

````python
# Check out the output
print(f"board: {board}")
print(f"board[0]: {board[0]}")
print(f"board[0][0]: {board[0][0]}")
````

````python
# Let's assign some value and see the output
board[0][0] = 'X'
print(board)
````

Something seems wrong. How can we make it right?

````python
board = [['']*3 for _ in range(3)]
board[0][0] = "X"
print(board)
````
